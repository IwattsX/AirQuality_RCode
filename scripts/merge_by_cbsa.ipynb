```python
import pandas as pd
```

```python
# Read datasets

df = pd.DataFrame()

for i in range(1980, 2025):
    filename = f"daily_aqi_by_cbsa_{i}.csv"
    tmp = pd.read_csv(f"data/{filename}", low_memory=False)
    df = pd.concat([df, tmp], axis=0) 
```

```python
df.head()
```

```python
# filter datasets by State and County
philly_df = df[df["CBSA"].str.contains("Philadelphia")]
philly_df.head()
```

```python
# write the merged dataset to a file
philly_df.to_csv("philly_csba.csv")
```
